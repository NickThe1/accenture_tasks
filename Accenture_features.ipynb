## Карта интенсивности

```python
df_corr = df.corr()

plt.figure(figsize = (16,5))
sns.heatmap(df_corr, annot=True, cmap="crest");
```

## Выводы

- Имеется положительная корреляция между 10 и (0, 1, 3, 4) переменными
- Корреляций между зависимыми переменными нет
